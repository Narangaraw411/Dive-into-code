[Problem 5] Python functionalization Please refer to the template to create a Python function that summarizes the problems 1 to 3.

Then, calculate the gradient of the following three equations at 0.1 intervals and output it with ndarray.

```python
import numpy as np
def compute_gradient(func, x=(-50, 50.1, 0.1)):
  x = np.arange(x[0],x[1],x[2])
  y = func(x)
  array_xy = np.stack([x,y], axis=1)
  xGrad = x[1:] - x[:-1]
  yGrad = y[1:] - y[:-1]
  gradient = yGrad/xGrad
  return array_xy, gradient
def function1(x):
  array_y = x**2
  return array_y

def function2(x):
  array_y = 2*x**2 + 2**x
  return array_y

def function3(x):
  array_y = np.sin(x**0.5)
  return array_y

array_xy, gradient = compute_gradient(function3)
#print(array_xy)
print(gradient)
import matplotlib.pyplot as plt

plt.plot(array_xy[:1000,0],gradient)

```
